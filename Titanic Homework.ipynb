```python
import pandas as pd

Location = "titanic3.xls"
df = pd.read_excel(Location)
df.head()
```

```python
df.describe()
```

```python
## sibsp is number of siblings or spouses each passenger had on board 
##   (same generation)
## parch is number of parents or children each passenger had on board 
##   (different generations)
```

```python
total_passengers = len(df)
total_women = len(df[df['sex'] == 'female'])
total_men = len(df[df['sex'] == 'male'])
total_survived = len(df[df['survived'] == 1])
total_firstclass = len(df[df['pclass'] == 1])
total_secondclass = len(df[df['pclass'] == 2])
total_thirdclass = len(df[df['pclass'] == 3])
total_no_sibsp = len(df[df['sibsp'] == 0])
total_one_sibsp = len(df[df['sibsp'] == 1])
total_twoplus_sibsp = len(df[df['sibsp'] > 1])

womensurvived = len(df[(df['sex'] == 'female') & (df['survived'] ==1)])
mensurvived = len(df[(df['sex'] == 'male') & (df['survived'] ==1)])
firstclasssurvived = len(df[(df['pclass'] ==1) & (df['survived'] ==1)])
secondclasssurvived = len(df[(df['pclass'] ==2) & (df['survived'] ==1)])
thirdclasssurvived = len(df[(df['pclass'] ==3) & (df['survived'] ==1)])
no_sibspsurvived = len(df[(df['sibsp'] ==0) & (df['survived'] ==1)])
one_sibspsurvived = len(df[(df['sibsp'] ==1) & (df['survived'] ==1)])
twoplus_sibspsurvived = len(df[(df['sibsp'] >1) & (df['survived'] ==1)])

p_womensurvived = womensurvived/total_survived
p_mensurvived = mensurvived/total_survived
p_firstclasssurvived = firstclasssurvived/total_survived
p_secondclasssurvived = secondclasssurvived/total_survived
p_thirdclasssurvived = thirdclasssurvived/total_survived
p_no_sibspsurvived = no_sibspsurvived/total_survived
p_one_sibspsurvived = one_sibspsurvived/total_survived
p_twoplus_sibspsurvived = twoplus_sibspsurvived/total_survived
p_survived = total_survived/total_passengers
```

```python
## Assignment is to determine which of the following had the greatest
##    likelihood of survival:
## 1-A male in second class with no siblings or spouse on board;
## 2-A female in third class with one sibling or spouse on board; or 
## 3-A male in first class with two or more siblings and/or spouse on board.
```

```python
Bayesnum_survival_if_male_secondclass_no_sibsp = p_mensurvived*p_secondclasssurvived*p_no_sibspsurvived*p_survived
Bayesnum_survival_if_female_thirdclass_one_sibsp = p_womensurvived*p_thirdclasssurvived*p_one_sibspsurvived*p_survived
Bayesnum_survival_if_male_firstclass_twoplus_sibsp = p_mensurvived*p_firstclasssurvived*p_twoplus_sibspsurvived*p_survived
```

```python
print(Bayesnum_survival_if_male_secondclass_no_sibsp)
```

```python
print(Bayesnum_survival_if_female_thirdclass_one_sibsp)
```

```python
print(Bayesnum_survival_if_male_firstclass_twoplus_sibsp)
```

```python
## Comparing the values of the Bayes numerators shows that, of the three:

## A female in third class with one sibling or spouse had the greatest 
##    likelihood of survival.
## A male in second class with no sibling or spouse had the next highest 
##    likelihood of survival.
## A male in first class with two or more siblings and/or spouse had the 
##    lowest likelihood of survival.

## Note that the values printed are not the actual probabilities, as only the 
##    Bayes numerators are needed for comparison purposes because the 
##    denominators are identical.
```

```python
## For information and additional analysis in the output:

print("Total passengers: ",total_passengers," women ",total_women," men ",total_men)
print("Total survivors:  ", total_survived,"  women ",womensurvived," men ",mensurvived)
print("% who survived:   ",int(100*total_survived/total_passengers),"%  women ",int(100*womensurvived/total_women),"% men ",int(100*mensurvived/total_men),"%")
print(" ")
print("Passengers in:  first class ",total_firstclass," second class ",total_secondclass," third class ",total_thirdclass)
print("Survivors in:   first class ",firstclasssurvived," second class ",secondclasssurvived," third class ",thirdclasssurvived)
print("% who survived: first class ",int(100*firstclasssurvived/total_firstclass),"% second class ",int(100*secondclasssurvived/total_secondclass),"% third class ",int(100*thirdclasssurvived/total_thirdclass),"%")
print(" ")
print("Passengers with: no sibsp ",total_no_sibsp," one sibsp ",total_one_sibsp," two+ sibsp ",total_twoplus_sibsp)
print("Survivors with:  no sibsp ",no_sibspsurvived," one sibsp ",one_sibspsurvived," two+ sibsp ",twoplus_sibspsurvived)
print("% who survived:  no sibsp ",int(100*no_sibspsurvived/total_no_sibsp),"% one sibsp ",int(100*one_sibspsurvived/total_one_sibsp),"% two+ sibsp ",int(100*twoplus_sibspsurvived/total_twoplus_sibsp),"%")
print(" ")
x1 = len(df[(df['sex'] == 'male') & (df['pclass'] ==2) & (df['sibsp'] ==0)])
x1s = len(df[(df['sex'] == 'male') & (df['pclass'] ==2) & (df['sibsp'] ==0) & (df['survived'] ==1)])
print("Total male passengers in second class with no siblings or spouse: ",x1)
print("Total who survived:                                                ",x1s)
print("% who survived:                                                    ",int(100*x1s/x1),"%")
print(" ")
x2 = len(df[(df['sex'] == 'female') & (df['pclass'] ==3) & (df['sibsp'] ==1)])
x2s = len(df[(df['sex'] == 'female') & (df['pclass'] ==3) & (df['sibsp'] ==1) & (df['survived'] ==1)])
print("Total female passengers in third class with one sibling or spouse: ",x2)
print("Total who survived:                                                ",x2s)
print("% who survived:                                                    ",int(100*x2s/x2),"%")
print(" ")
x3 = len(df[(df['sex'] == 'male') & (df['pclass'] ==1) & (df['sibsp'] >1)])
x3s = len(df[(df['sex'] == 'male') & (df['pclass'] ==1) & (df['sibsp'] >1) & (df['survived'] ==1)])
print("Total male passengers in first class with two+ siblings and/or spouse: ",x3)
print("Total who survived:                                                    ",x3s)
print("% who survived:                                                       ",int(100*x3s/x3),"%")
```
